```python
from sympy import Matrix
from sympy.abc import k,j,B,N,X,x,F,q,g,o,z
import random

def Rule():
    print("""
    Bonjour !
    Voici les règles du démineur !
    Ce jeu est composé d'une grille d'une dimension de votre choix,
    pour construire votre grille, vous devrez utiliser la commande Dem(ligne,colonne).
    Une fois que vous avez construit la grille, des bombes seront dissimulées aléatoirement
    et ce sera à vous à les trouver !
    Le nombre de bombes est entre 1 et un certains nombre qui vous sera dit et vous sera demander.
             
    Comment ?
    La grille est dite face cachée, pour voir ce qu'il y a derrière une case,
    il va falloir faire la commande Show(ligne,colonne).
    Dans la grille, il va il y avoir des bombes mais aussi des numéros qui vont vous indiquer combien de bombes
    il y a au alentour de cette case (en haut, bas, droite, gauche et en diagonale).
    Avec vous, vous aurez un certain nombre de drapeaux égal au nombre de bombes.
    Les drapeaux devront être placé sur les bombes avec la commande Flag(ligne,colonne)
    
    Et après ?
    Une fois que vous avez mis tous les drapeaux et que vous avez découvert toutes les cases,
    vous pouvez demander à la machine si vous avez gagné ou non avec la commande Win().
    
    PS: Si besoin d'aide, utilisé la commande Help().
    Attention : Les lignes et colonnes sont comptée à partir de 0
    """)

def Help():
    print(
    """    Découvrir une case, Show(ligne,colonne)
    Mettre/retirer un drapeau, Flag(ligne,colonne)
    Attention : Les lignes et colonnes sont comptée à partir de 0""")
    
    
def Dem(k=8,j=8):
    global o
    global z
    o=k
    z=j
    while j<8:
        print("Le nombre de colonnes est invalide")
        j=int(input("Combien de colonnes voullez-vous? "))
        
    while k<8:
        print("Le nombre de lignes est invalide")
        k=int(input("Combien de lignes voullez-vous? "))
        
    global X
    global N
    global nb
    global M
    M=Matrix([[0]*j]*k)
    X=Matrix([[" "]*j]*k)
    print("Le nombre maximum de bombes est de :",int((j*k)/5))
    N=int(input("Combien de bombes voullez-vous? "))
    nb=N
    
    while N>int(((j*k)/5)) or N<1:
        print("Nombre de bombes invalides")
        print("Le nombre maximum de bombes est de :",int((j*k)/5))
        N=int(input("Combien de bombes voullez-vous? "))
    while list(M).count(B)<N:
         M[random.randint(0,k-1),random.randint(0,j-1)]=B

    if M[0]==B:
        for y in [1,j,j+1]:
            if M[y]!=B:
                M[y]+=1

    if M[j-1]==B:
        for y in [j-2,(2*j)-1,(2*j)-2]:
            if M[y]!=B:
                M[y]+=1

    if M[(k*j)-j]==B:
        for y in [j-1,2*j,(2*j)-1]:
            if M[(k*j)-y]!=B:
                M[(k*j)-y]+=1

    if M[(k*j)-1]==B:
        for y in [2,j+2,j+1]:
            if M[(k*j)-y]!=B:
                M[(k*j)-y]+=1

    for C1 in range(1,j-1):
        if (M[C1]==B):
            for y in [-1,1,j,j-1,j+1]:
                if M[C1+y]!=B:
                    M[C1+y]+=1

    for C2 in range((k*j)-j+1,(k*j)-1):
        if (M[C2]==B):
            for y in [-1,1,-j,-j-1,-j+1]:
                if M[C2+y]!=B:
                    M[C2+y]+=1

    for C3 in range(j,(k*j)-2*j+2,j):
        if (M[C3]==B):
            for y in [-j,-j+1,1,j,j+1]:
                if M[C3+y]!=B:
                    M[C3+y]+=1

    for C4 in range(2*j-1,(k*j)-1,j):
        if (M[C4]==B):
            for y in [-j,-j-1,-1,j,j-1]:
                if M[C4+y]!=B:
                    M[C4+y]+=1

    for C in range(1,k-1):
        for L in range(1,j-1):
            if M[C,L]==B:
                for y in [-1,0,1]:
                    for g in [-1,0,1]:
                        if M[C+g,L+y]!=B:
                            M[C+g,L+y]+=1
    print("Bonne chance !")
    return X

def Flag(q,g):
    global N
    if N>0:
        if X[q,g]==x:
            X[q,g]=F
            N-=1
            print("Drapeaux :",N)
        elif X[q,g]==F:
            X[q,g]=x
            N+=1
            print("Drapeaux :",N)
    elif X[q,g]==F:
        X[q,g]=x
        N+=1
        print("Drapeaux :",N)
    elif N==0:
        print("Vous avez déjà mis tous vos drapeaux !")
    return X

def Show(q,g):
    if X[q,g]==F:
        print("Cette case a été marquée par un drapeau !")
    elif X[q,g]==x:
        if M[q,g]==B:
            print("BOOOOOOOOOOOM")
            return restart()
        if M[q,g]==0:
            X[q,g]=0
            if q==0 and g==0:
                Show(0,1)
                Show(1,1)
                Show(1,0)
            if q==o-1 and g==0:
                Show(o-1,1)
                Show(o-2,1)
                Show(o-2,0)
            if q==0 and g==z-1:
                Show(0,z-2)
                Show(1,z-2)
                Show(1,z-1)
            if q==o-1 and g==z-1:
                Show(o-1,z-2)
                Show(o-2,z-1)
                Show(o-2,z-2)
            for i in range(1,z-1):
                if q==0 and g==i:
                    for y in [i,i+1,i-1]:
                        if X[1,y]==x:
                            Show(1,y)
                    for y in [i-1,i+1]:
                        if X[0,y]==x:
                            Show(0,y)
                if q==o-1 and g==i:
                    for y in [i,i+1,i-1]:
                        if X[o-2,y]==x:
                            Show(o-2,y)
                    for y in [i-1,i+1]:
                        if X[o-1,y]==x:
                            Show(o-1,y)
            for i in range(1,o-1):
                if q==i and g==0:
                    for y in [i-1,i+1]:
                        if X[y,0]==x:
                            Show(y,0)
                    for y in [i-1,i,i+1]:
                        if X[y,1]==x:
                            Show(y,1)
                if q==i and g==z-1:
                    for y in [i-1,i,i+1]:
                        if X[y,z-2]==x:
                            Show(y,z-2)
                    for y in [i-1,i+1]:
                        if X[y,z-1]==x:
                            Show(y,z-1)
            for i in range(1,o-1):
                for l in range(1,z-1):
                    if q==i and g==l:
                        for y in [-1,0,1]:
                            for g in [-1,0,1]:
                                if X[i+y,l+g]==x:
                                    Show(i+y,l+g)
            return X
        else:
            for i in range(1,9):
                if M[q,g]==i:
                    X[q,g]=i
    return X

def Win():
    if list(X).count(F)==list(M).count(B) and list(X).count(x)==0 :
        print("BRAVO VOUS AVEZ GAGNE !!!!")
        X
        restart()
    else:
        print("Pas pour l'instant...")
        return X

def Debrief():
    print("Vous avez choisi une grille de",o,"sur",z)
    print("Vous avez choisi",nb,"bombes")
    print("Il vous reste",N,"Drapeaux")

def restart():
    r=str(input("Voulez-vous recommencer ? (y/n)"))
    if r=="y":
        Dem(int(input("Combien de lignes voulez-vous? ")),int(input("Combien de colonnes voulez-vous? ")))
    elif r=="n":
        print("Merci d'avoir jouer !")
    else:
        print("Votre réponse est incompatible")
        restart()
Rule()
```
