```python
#Importer la définition array est juste un aspect esthétique pour aligner les lignes et colonnes de la grille.
from numpy import array
#La définition randint sert à importer des nombres aléatoires pour le placement des bombes dans la grille.
from random import randint

#n=nbr ligne, m=nbr colonne, k=nbr bombe



#Commande de la matrice
def matrice(n,m):
#Nous créons une nouvelle définition nommée matrice et qui à besoin de 2 constantes non prédéfinies pour pouvoir être exécutée.
    matrice =[]
#On a créé une liste vide nommée matrice.
    for i in range(0,n):
#On a créé une liste vide nommée l.
        l=[]
        for j in range(0,m):
#On a remplit cette liste l par des "0".
            l.append(0)
#On a intercalé dans la liste matrice la liste l remplit de "0".
        matrice.append(l)
    return matrice
#Cette boucle va être répétée n fois, ainsi que la boucle interne qui va permettre d’ajouter des éléments "0" dans l, ce qui va former les lignes (n) et les colonnes (m).
#Et pour finir elle va ressortir "matrice" qui va être notre grille.



#Génération des bombes
def bombe(n,m,k):
#Nous créons une nouvelle définition nommée bombe et qui à besoin de 3 constantes non prédéfinies pour pouvoir être exécutée.
    if k <=int(n*m/5):
#La première chose que la fonction va faire, c’est de calculer si le nombre de bombes "k" est trop haut par rapport à la limite.
        matrix = matrice(n,m)
#Ici, nous mettons la fonction matrice en route à l’intérieur d’une autre tout en la nommant "matrix".
        for i in range(0,k):
            a=randint(0,n-1)
            b=randint(0,m-1)
            while matrix[a][b]=="B":
                a=randint(0,n-1)
                b=randint(0,m-1)
            matrix[a][b]="B"
#Pour cette boucle, elle va être répétée k fois pour le nombre de bombe.
#Les coordonnées aléatoires vont être choisies entre 0 et n-1 car python va compter à partir de 0. De plus, on a rajouté une boucle qui modifie les coordonnées d’une bombe si celle-ci était positionnée sur une case où il y avait déjà une bombe.
    return matrix
#Et pour finir, nous retournons "matrix".



#Création de la matrice que verra le joueur
def Player(n,m):
#Nous créons une nouvelle définition nommée Player et qui à besoin de 2 constantes non prédéfinies pour pouvoir être exécutée.
    matrice =[]
    for i in range(0,n):
        l=[]
        for j in range(0,m):
            l.append("X")
        matrice.append(l)
#Nous procédons de la même manière que pour la fonction matrice.
    return matrice
#C’est exactement la même fonction que matrice mais nous mettons "X" à la place de "0".



#comptage de bombes aux alentours d'une case
def number (matrice,n,m):
#Nous créons une nouvelle définition nommée number et qui à besoin de 3 constantes non prédéfinies pour pouvoir être exécutée.
    for i in range(1,n-1):
        for j in range(1,m-1):
#Ici nous formons un balayage de toutes les cases sauf les extrémités. Ces deux boucles vont former toutes les coordonnées internes de la liste de listes.
            if matrice[i][j]=="B":
#Nous avons une condition qui nous dit simplement, si sur cette case-là , il y a une bombe, alors faites ceci : elle regarde si les cases aux alentours de cette derniére n'ont pas de bombes.
                if matrice[i-1][j-1] != "B":
                    matrice[i-1][j-1]+=1
#Et nous retrouvons encore une condition qui nous dit, si la case à droite de celle où il y a une bombe, n’est pas une bombe, alors ajoute 1 à cette case. Et on va pouvoir retrouver exactement la même chose sur les cases suivantes qui vont être le contour de la case ayant une bombe.

                if matrice[i][j-1]!="B":
                    matrice[i][j-1]+=1
                if matrice[i+1][j-1]!= "B":
                    matrice[i+1][j-1]+=1
                if matrice[i+1][j]!= "B":
                    matrice[i+1][j]+=1
                if matrice[i+1][j+1]!= "B":
                    matrice[i+1][j+1]+=1
                if matrice[i][j+1]!= "B":
                    matrice[i][j+1]+=1
                if matrice[i-1][j+1]!= "B":
                    matrice[i-1][j+1]+=1
                if matrice[i-1][j]!= "B":
                    matrice[i-1][j]+=1
#Pour cette condition (hors de la boucle du dessus), elle nous dit, si la case tout en haut à droite est une bombe, alors faite la suite. 
    if matrice[0][0] == "B":
#On retrouve encore une condition qui dit, si la case à droite du coin n’est pas une bombe, alors on y ajoute 1.
#Par rapport à la boucle au dessus,ici on a des cases en moins qui ne sont pas à prendre en compte car elles n'existent pas.
        if matrice[0][1]!="B":
            matrice[0][1]+=1
        if matrice[1][0] !="B":
            matrice[1][0]+=1
        if matrice[1][1] != "B":
            matrice[1][1] +=1

    if matrice[0][m-1]=="B":
#Cette condition est la même adaptée au coin droit en haut.
        if matrice[0][m-2]!="B":
            matrice[0][m-2]+=1
        if matrice[1][m-1]!="B":
            matrice[1][m-1]+=1
        if matrice[1][m-2]!="B":
            matrice[1][m-2]+=1

    if matrice[n-1][0]=="B":
#Cette condition est la même adaptée au coin gauche en bas.
        if matrice[n-1][1]!="B":
            matrice[n-1][1]+=1
        if matrice[n-2][1]!="B":
            matrice[n-2][1]+=1
        if matrice[n-2][0]!="B":
            matrice[n-2][0]+=1

    if matrice[n-1][m-1]=="B":
#Cette condition est la même adaptée au coin droite en bas.
        if matrice[n-1][m-2]!="B":
            matrice[n-1][m-2]+=1
        if matrice[n-2][m-1]!="B":
            matrice[n-2][m-1]+=1
        if matrice[n-2][m-2]!="B":
            matrice[n-2][m-2]+=1

    for k in range(1,m-1):
        if matrice[0][k]=="B":
#La boucle va faire un balayage de toutes les cases des lignes sans extrémités.
#La condition va dire si la case est une bombe alors fait la chose suivante.

            if matrice[0][k-1]!="B":
                matrice[0][k-1]+=1
            if matrice[0][k+1]!="B":
                matrice[0][k+1]+=1
            if matrice[1][k-1] !="B":
                matrice[1][k-1]+=1
            if matrice[1][k]!="B":
                matrice[1][k]+=1
            if matrice[1][k+1]!="B":
                matrice[1][k+1]+=1
                
#On va retrouver 5 conditions qui veulent dire la même chose, si la case à droite n’est pas une bombe, alors on y ajoute 1. De même pour les cases qui entourent celle-ci.
        if matrice[n-1][k]=="B":
#Dans la même boucle, on fait la même chose que ci-dessus mais avec un balayage sur la dernière ligne sans compter les extrémités.
            if matrice[n-1][k-1]!="B":
                matrice[n-1][k-1]+=1
            if matrice[n-2][k-1]!="B":
                matrice[n-2][k-1]+=1
            if matrice[n-2][k] !="B":
                matrice[n-2][k]+=1
            if matrice[n-2][k+1] !="B": 
                 matrice[n-2][k+1]+=1
            if matrice[n-1][k+1]!="B":
                 matrice[n-1][k+1]+=1
                    
#Même explication que ci-dessus.        
    for k in range(1,n-1):
        if matrice[k][0]=="B":
#La boucle va faire un balayage de toutes les cases des colonnes sans les extrémités.

#La condition va dire si la case de la première colonne est une bombe alors fait la chose suivante.
            if matrice[k-1][0]!="B":
                matrice[k-1][0]+=1
            if matrice[k-1][1]!="B":
                matrice[k-1][1]+=1
            if matrice[k][1]!="B":
                matrice[k][1]+=1
            if matrice[k+1][1]!="B":
                matrice[k+1][1]+=1
            if matrice[k+1][0]!="B":
                matrice[k+1][0]+=1
                
#Même explication que les lignes mais en première colonne.                    
        if matrice[k][m-1]=="B":
#La condition va dire si la case de la dernière colonne est une bombe alors appliquer ses conditions.
            if matrice[k-1][m-1]!="B":
                matrice[k-1][m-1]+=1
            if matrice[k-1][m-2]!="B":
                matrice[k-1][m-2]+=1
            if matrice[k][m-2]!="B":
                matrice[k][m-2]+=1
            if matrice[k+1][m-2]!="B":
                matrice[k+1][m-2]+=1
            if matrice[k+1][m-1]!="B":
                matrice[k+1][m-1]+=1
                
#Même explication que les lignes mais en dernière colonne.
    return matrice 
##Pour finir, nous retournons la matrice initier dans le processus de comptage de bombes.



#Fusion de tout+limite
#Vérification de tous les données 
def D(n,m,k):
#Nous créons une nouvelle définition nommée D et qui à besoin de 3 constantes non prédéfinies pour pouvoir être exécutée.
    global i
    global j
    global mb
#Les trois global vont être des valeurs qui vont être enregistrés partout dans la console et non seulement dans cette définition.
    while m<8 or m!=int(m):
        print("Le nombre de colonnes est invalide")
        m=int(input("Combien de colonnes voullez-vous? "))
#Ici, nous avons une boucle qui va se mettre en route si l'utilisateur à mis un nombre de colonnes en dessous de 8. Si celui-ci en a mis trop peu, nous allons demander combien de collones le joueur voudrait jusqu'à ce qu'il en mette un bon.
    
    while n<8 or n!=int(n):
        print("Le nombre de lignes est invalide")
        n=int(input("Combien de lignes voullez-vous? "))
#De même pour les lignes. Nous avons une boucle qui va se mettre en route si l'utilisateur à mis un nombre de lignes en dessous de 8. Si celui-ci en a mis trop peu, nous allons demander combien de lignes le joueur voudrait jusqu'à ce qu'il en mette un bon.

    while (k>int(((n*m)/5)) or k<1)or k!=int(k):
        print("Nombre de bombes invalide")
        print("Le nombre maximum de bombes est de :",int((n*m)/5))
        k=int(input("Combien de bombes voullez-vous? "))
#De même pour les bombes.    
    j=int(m)
    i=int(n)  
    mb=int(k)
#Ici, nous créons des copies de toutes les données de la grille (lien avec global), pour les avoirs partout dans la console.    
    global M
    M=number(bombe(n,m,k),n,m)
#Ici encore une fois nous faisons une copie de la grille qui va être enclenchée si les while ont tous été passés avec succès.
    global C
    C=Player(n,m)
#De même avec la grille que le joueur va voir.
    return print(array(C))
#Et on affiche la grille du joueur.



#Placement des drapeaux 
def Flag(x,y):
#Nous créons une nouvelle définition nommée Flag et qui à besoin de 2 constantes non prédéfinies pour pouvoir être exécutée.
#x=position de la ligne et y=position de la colonne
    global mb
#Elle se rappelle le nombre de bombes dans la grille (la copie).
    if x<i and x>=0 and y<j and y>=0 :
#La condition va demander si les coordonnées sont dans la grille, si oui voici les conditions. 
        if mb>0:
            if C[x][y]=="X":
                C[x][y]="F"
                mb-=1
                print("Il vous reste",mb,"drapeau(x)")
#Si le nombre de bombes est supérieur à 0, si la case n’est pas dévoilée, elle va placer un drapeau et va retirer un drapeau au compteur et prévenir de ce nombre.
            elif C[x][y]=="F":
                C[x][y]="X"
                mb+=1
                print("Il vous reste",mb,"drapeau(x)")
#Si le nombre de bombes est supérieur à 0, si la case n’est pas dévoilée, elle va retirer un drapeau et va ajouter un drapeau au compteur et prévenir de ce nombre.
            else:
                print("Cette case est déjà découverte !")
#Si la case est dévoilée et non marquée alors la machine indique que la case à déjà été découverte.
        elif C[x][y]=="F":
            C[x][y]="X"
            mb+=1
            print("Il vous reste",mb,"drapeau(x)")
#Si le nombre de bombes est égale à 0, si la case a été marquée, elle va retirer un drapeau et va ajouter un drapeau au compteur et prévenir de ce nombre.
        elif mb==0 or C[x][y]!="F":
            print("Vous n'avez plus de drapeau")
#Si le nombre de drapeaux est nul, alors elle vous avertira.


    else:
        print("La position du drapeau n'est pas compatible avec la grille")

    print(array(C))
#Si non (pour la position), elle va prévenir que le drapeau n’est pas plaçable et elle imprime C.

    return win()
#La fonction win va se contrôler à chaque drapeaux mis s’il a fini la grille.



#Dévoilement d'une case
def S(n,m):
#Nous créons une nouvelle fonction nommée S, qui a besoin de 2 coordonnées fournies par la fonction Show.
    if C[n][m]=="F":
        print("Cette case a été marquée par un drapeau !")
#Pour la première condition, si les coordonnées correspondent à un drapeau (dans la matrice du joueur) alors elle prévient que cette case à été marquée.
    elif C[n][m]!="F" and C[n][m]!="X":
        print("Cette case a déjà été dévoilée !")
#Pour cette condition, si, dans la matrice du joueur, les coordonnées ne correspondent pas à un drapeau et pas à une case non dévoilée, elle prévient que la case à déjà été dévoilée.
    elif C[n][m]=="X":
        if M[n][m]=="B":
            print("BOOOOOOOOOOOM")
            lost()
            return restart()
#Ici, nous avons une condition dans une condition, pour la première, si la case du joueur n’est pas dévoilée, alors s' il s’agit d’une bombe, alors le joueur à perdu, la machine montre la matrice cachée et actionne la fonction restart.
        elif M[n][m]==0:
            C[n][m]=0
#Ici, cette condition va voir si dans la matrice M (celle cachée du joueur) au coordonnée n,m ça correspond à 0, alors il va définir sur C en n,m un 0 et va actionner toutes les conditions suivantes.
            if n==0 and m==0:
                if C[0][1]=="X":
                    S(0,1)
                if C[1][1]=="X":
                    S(1,1)
                if C[1][0]=="X":
                    S(1,0)
#Si le coin en haut à gauche est un zéro alors si les cases autours sont non découvertes alors elle va les développer.
            if n==i-1 and m==0:
                if C[i-1][1]=="X":
                    S(i-1,1)
                if C[i-2][1]=="X":
                    S(i-2,1)
                if C[i-2][0]=="X":
                    S(i-2,0)
#De même.
            if n==0 and m==j-1:
                if C[0][j-2]=="X":
                    S(0,j-2)
                if C[1][j-2]=="X":
                    S(1,j-2)
                if C[1][j-1]=="X":
                    S(1,j-1)
#De même.
            if n==i-1 and m==j-1:
                if C[i-1][j-2]=="X":
                    S(i-1,j-2)
                if C[i-2][j-1]=="X":    
                    S(i-2,j-1)
                if C[i-2][j-2]=="X":    
                    S(i-2,j-2)
#De même.
            for r in range(1,j-1):
                if n==0 and m==r:
                    if C[0][r-1]=="X":
                        S(0,r-1)
                    if C[0][r+1]=="X":
                        S(0,r+1)
                    if C[1][r]=="X":
                        S(1,r)
                    if C[1][r+1]=="X":
                        S(1,r+1)
                    if C[1][r-1]=="X":
                        S(1,r-1)
#Cette boucle va répéter ces conditions pour développer les cases autours des cases balayées sur le côté gauche.
                if n==i-1 and m==r:
                    if C[i-1][r-1]=="X":
                        S(i-1,r-1)
                    if C[i-1][r+1]=="X":
                        S(i-1,r+1)
                    if C[i-2][r]=="X":
                        S(i-2,r)
                    if C[i-2][r+1]=="X":
                        S(i-2,r+1)
                    if C[i-2][r-1]=="X":
                        S(i-2,r-1)
#De même pour le côté droit.
            for r in range(1,i-1):
                if n==r and m==0:
                    if C[r-1][0]=="X":
                        S(r-1,0)
                    if C[r-1][1]=="X":
                        S(r-1,1)
                    if C[r][1]=="X":
                        S(r,1)
                    if C[r+1][1]=="X":
                        S(r+1,1)
                    if C[r+1][0]=="X":
                        S(r+1,0)
                if n==r and m==j-1:
                    if C[r-1][j-1]=="X":
                        S(r-1,j-1)
                    if C[r-1][j-2]=="X":
                        S(r-1,j-2)
                    if C[r][j-2]=="X":
                        S(r,j-2)
                    if C[r+1][j-2]=="X":
                        S(r+1,j-2)
                    if C[r+1][j-1]=="X":
                        S(r+1,j-1)
#De même que pour les côtés mais cette fois sur les côtés du haut et du bas.
            for r in range(1,i-1):
                for l in range(1,j-1):
                    if n==r and m==l:
                        if C[r][l+1]=="X":
                            S(r,l+1)
                        if C[r+1][l+1]=="X":
                            S(r+1,l+1)
                        if C[r+1][l]=="X":
                            S(r+1,l)
                        if C[r+1][l-1]=="X":
                            S(r+1,l-1)
                        if C[r][l-1]=="X":
                            S(r,l-1)
                        if C[r-1][l-1]=="X":
                            S(r-1,l-1)
                        if C[r-1][l]=="X":
                            S(r-1,l)
                        if C[r-1][l+1]=="X":
                            S(r-1,l+1)
#Ici, nous avons une double boucle qui va balayer toutes les coordonnées centrale de la matrice (sans les coins et bords).
        else:
            for r in range(1,9):
                if M[n][m]==r:
                    C[n][m]=r
#Pour finir, si ce n’est pas une bombe ni un 0, cela veut dire que c’est un nombre de 1 à 9, donc la boucle va répéter la condition, si les coordonnées.


                    
#Dévoilement d'une case                    
def Show(n,m):
#Nous créons une nouvelle définition nommée Show qui a besoin de 2 coordonnées pour pouvoir être exécutée.
    S(n,m)
#Les coordonnées vont être implantée dans la fonction S.
    if M[n][m]!="B" or C[n][m]=="F":
#La condition ci-dessus veut dire que si les coordonnées ne sont pas sur une bombe (dans la matrice totalement découverte) ou si la case de la grille choisie est marquée par un drapeau.
        print(array(C))
    win()
#A chaque fois que le joueur va poser un drapeaux, la fonction win va tester.



#Recommencer le jeu
def restart():
#Nous créons une nouvelle définition nommée restart et qui à besoin d’aucune constante pour pouvoir être exécutée.
    r=str(input("Voulez-vous recommencer ? (y/n)"))
#y=oui et n=non
    if r=="y":
        D(int(input("Combien de lignes voulez-vous? ")),int(input("Combien de colonnes voulez-vous? ")),int(input("Combien de bombes voulez-vous? ")))
#La machine demande si le joueur voudrait recommencer ou non, si oui elle va demander les lignes, les colonnes et les bombes voulues. 
    elif r=="n":
        print("Merci d'avoir jouer !")
#Sinon elle écrit un petit message.
    else:
        print("Votre réponse est incompatible")
        restart()
#Si une autre réponse à été mise, que y ou n, la machine va nous mettre un message d'erreur et va recommencer.



#Gagner le jeu
def win():
#Nous créons une nouvelle définition nommée Win et qui à besoin d’aucune constante pour pouvoir être exécutée.
    p=0
#Nous assignons un compteur 0 nommé p.
    if mb==0 :
        for u in range(0,i):
            p+=C[u].count("X")
#Pour cette condition, elle va voir si le nombre de drapeaux est à 0, si oui, elle va compter le nombre de cases non dévoilées.
        if p==0:
            print("Félicitation, vous avez gagné !")
            restart()
#Si le compteur p (de case non dévoilée) est à 0, la machine nous dit félicitation, nous montre la grille et enclenche restart.



#L'aide reprenant des fonctions    
def help():
#Nous créons une nouvelle définition nommée help et qui à besoin d’aucune constante pour pouvoir être exécutée.
    print("Flag(x,y) : place un drapeau a la position (x,y) ")
    print("Show(x,y) : découvre la case (x,y)")
    print("restart() : recommence le jeu")
#La fonction créée ci-dessus sert à afficher les commandes que le joueur peut utiliser.
  
    
    
#Commencer le jeu.  
def start():
#Nous créons une nouvelle définition nommée start et qui à besoin d’aucune constante pour pouvoir être exécutée.
    print("Bienvenue dans le jeu Démineur !")
    D(int(input("Combien de lignes voulez-vous? ")),int(input("Combien de colonnes voulez-vous? ")),int(input("Combien de bombes voulez-vous? ")))
    help()
#Cette fonction reprenant la fonction D va nous demander de choisir le nombre de lignes, colonnes et bombes qu'on veut pour notre grille et applique help().
 


#Emplacement des bombes
def lost():
#Nous créons une nouvelle définition nommée lost et qui à besoin d’aucune constante pour pouvoir être exécutée.
    for x in range(i):
        for y in range(j):
            if M[x][y]== "B":
                C[x][y]="B"
    print(array(C))
#On regarde sur chaque valeur de la matrice M si elle contient une bombe, si tel est le cas, on l'affiche sur la matrice C.


start()
```
